### Make the database

Next we want to make a database of UniProt/Swiss-prot

```python
!pwd
```

```python
!mkdir db
```

```python
!ls
```

```python
%%bash
curl ftp://ftp.uniprot.org/pub/databases/uniprot/current_release/knowledgebase/complete/uniprot_sprot.fasta.gz \
> ./db/uniprot_sprot.fasta.gz
```

```python
!ls ./db/
```

```python
%%bash
gunzip -k ./db/uniprot_sprot.fasta.gz
```

```python
!ls ./db/
```

Now we append the version number to the database using the program makeblastdb

```python
%%bash
/c/Users/FishLab3/Bioinfo-progs/blast-2.7.1+/bin/makeblastdb \
-in ./db/uniprot_sprot.fasta \
-dbtype prot \
-out ./db/uniprot_sprot_r2018_08
```

```python
!ls ./db/
```

### Get Query Sequence

```python
%%bash
curl http://eagle.fish.washington.edu/cnidarian/Ab_4denovo_CLC6_a.fa \
> ./query/Ab_4denovo_CLC6_a.fa
```

Preview the data

```python
!head ./query/Ab_4denovo_CLC6_a.fa
```

How many sequences? Use `grep` to count ">" because eahc contig starts with one

```python
%%bash
grep -c ">" ./query/Ab_4denovo_CLC6_a.fa
```

### Run Blast

```python
%%bash
/c/Users/FishLab3/Bioinfo-progs/blast-2.7.1+/bin/blastx \
-query ./query/Ab_4denovo_CLC6_a.fa \
-db ./db/uniprot_sprot_r2018_08 \
-out ./analyses/Ab_4-uniprot_blastx.tab \
-evalue 1E-20 \
-num_threads 4 \
-max_target_seqs 1 \
-outfmt 6
```

```python
!head ./analyses/Ab_4-uniprot_blastx.tab
```

So how many blast hits did we get? Use `wc` to determine number of lines

```python
%%bash
wc -l ./analyses/Ab_4-uniprot_blastx.tab
```
